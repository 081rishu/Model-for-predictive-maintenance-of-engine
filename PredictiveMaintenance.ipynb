## Step 4: Merging data to get RUL

#### Insights from train dataset
The first col has value from 1 - 100 and for each there is a range of values that indicates the engine and cylce 

```python
df_life_cycle = df.groupby([0]).agg({1:'max'})
df_test_life_cycle = df_test.groupby([0]).agg({1:'max'})
```

```python
df_life_cycle.head()
```

```python
df_test_life_cycle.head()
```

```python
df = df.merge(df_life_cycle, how='left', on=[0])
df_test = df_test.merge(df_test_life_cycle, how='left', on=[0])
```

```python
df.head()
```

```python
df_test.head()
```

there is a col in last named 1_y

```python
df['RUL'] = df['1_y'] - df['1_x']
df_test['RUL'] = df_test['1_y'] - df_test['1_x']
```

```python
df.drop(['1_y'], axis=1, inplace=True)
df_test.drop(['1_y'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df_test.head()
```

```python
boolean_columns = df.select_dtypes(include=['bool']).columns

df[boolean_columns] = df[boolean_columns].astype(int)
df_test[boolean_columns] = df_test[boolean_columns].astype(int)
```

## Step 5: Splitting data

```python
from sklearn.model_selection import train_test_split
```

```python
y = df['RUL']  
y_testData = df_test['RUL']
```

#### standardizing data

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.rename(columns={'1_x': '1'}, inplace=True)
df_test.rename(columns={'1_x': '1'}, inplace=True)

df.columns = df.columns.astype(str)
df_test.columns = df_test.columns.astype(str)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(df)
X_testData = scaler.fit_transform(df_test)
```

```python
X_df = pd.DataFrame(X, columns=df.columns) 
X_testData_df = pd.DataFrame(X_testData, columns=df.columns) 

X_df.info()
```

```python
X_testData_df.info()
```

```python
X_df.drop(columns=['RUL'], inplace=True)
X_testData_df.drop(columns=['RUL'], inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_df, y, test_size=0.2, random_state=100)
```

## Step 6: Model building

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred_train = model.predict(X_train)
```

```python
rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
r2_train = r2_score(y_train, y_pred_train)
print(f'RMSE on Training Data: {rmse_train}')
print(f'R2 Score on Training Data: {r2_train}')
```

```python
y_pred_test = model.predict(X_test)
```

```python
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
r2_test = r2_score(y_test, y_pred_test)
```

```python
print(f'RMSE on Testing Data: {rmse_test}')
print(f'R2 Score on Testing Data: {r2_test}')
```

```python
# Visualize Predictions
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_test, alpha=0.5)
plt.plot([0, max(y_test)], [0, max(y_test)], color='red', linestyle='--')
plt.xlabel('Actual RUL')
plt.ylabel('Predicted RUL')
plt.title('Actual vs Predicted RUL')
plt.show()
```

```python
y_pred_testData = model.predict(X_testData_df)
rmse_testData = np.sqrt(mean_squared_error(y_testData, y_pred_testData))
r2_testData = r2_score(y_testData, y_pred_testData)
print(f'RMSE on Testing Data: {rmse_testData}')
print(f'R2 Score on Testing Data: {r2_testData}')
```
